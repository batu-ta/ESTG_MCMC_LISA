```python
import numpy as np
import matplotlib.pyplot as plt
from getdist import plots, MCSamples

np.random.seed(42)
alpha_m_samples = np.random.normal(0.013, 0.037, 10000)
omega_cdm_samples = np.random.normal(0.12, 0.01, 10000)
h_samples = np.random.normal(0.67, 0.01, 10000)

samples = np.column_stack([alpha_m_samples, omega_cdm_samples, h_samples])
names = ['alpha_M', 'Omega_cdm', 'h']
labels = [r'\alpha_M', r'\Omega_{\rm cdm}', 'h']

chain = MCSamples(samples=samples, names=names, labels=labels)
g = plots.get_subplot_plotter()
g.triangle_plot(chain, filled=True)
plt.savefig('corner_plot.png')
plt.show()

print(f"alpha_M: {np.mean(alpha_m_samples):.3f} ± {np.std(alpha_m_samples):.3f}")
```
